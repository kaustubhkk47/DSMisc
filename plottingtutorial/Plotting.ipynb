### Useful arguements to plot
- color: r, g, b, y, k - black, w - white, c - cyan, m - magenta
- marker: o, v , ^, < , >, s , + , x , d, D
- linestyle: - , -- , -. , : 
- linewidth
- markersize  

fmt can also be used 

```python
plt.title('Line plot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()
```

```python
plt.title('Format test')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, 'go', label = 'linear')
plt.legend()
plt.show()
```

### Anatomy
The figure is the highest level object  
Axes can be added to the figure

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

#### The same as above but cleaner

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# or replace the three lines of code above by the following line: 
#fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])

# Show the plot
plt.show()
```

| Plot  | Meaning   |
|----|---------------|
| ax.bar()  | Vertical rectangles |
| ax.barh() | Horizontal rectangles |
| ax.axhline() | Horizontal line across axes |
| ax.vline() | Vertical line across axes |
| ax.fill() | Filled polygons |
| ax.fill_between() | Fill between y-values and 0 |
| ax.stackplot() | Stack plot |
| ax.hist() | Histogram |
| ax.boxplot() | Boxplot |
| ax.contour() | Contour plot |
|  |  |

```python
fig = plt.figure(figsize=(21,7))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

x = [ 0.  ,  0.25,  0.5 ,  0.75,  1.  ]
y = [ 0.  ,  1.25,  2.5 ,  3.75,  5.  ]
# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter(x,y)

ax3.legend('y')

# Show the plot
plt.show()
plt.savefig("foo.png")
```
